```python
class Supermarket:
    def __init__(self):
        self.nama_barang = []
        self.harga_barang = []
        self.jumlah_barang = []

    def tambah_barang(self, nama, harga, jumlah):
        self.nama_barang.append(nama)
        self.harga_barang.append(harga)
        self.jumlah_barang.append(jumlah)
        print("Barang ditambahkan ke keranjang.")

    def delete_item(self, nama):
        if nama in self.nama_barang:
            index = self.nama_barang.index(nama)
            del self.nama_barang[index]
            del self.harga_barang[index]
            del self.jumlah_barang[index]
            print("Barang dihapus dari keranjang.")
        else:
            print("Barang tidak ditemukan di keranjang.")

    def reset_transaction(self):
        self.nama_barang = []
        self.harga_barang = []
        self.jumlah_barang = []
        print("Transaksi telah direset.")

    def check_keranjang(self):
        print("Isi Keranjang Belanja:")
        for i in range(len(self.nama_barang)):
            print(f"Barang: {self.nama_barang[i]}, Harga: {self.harga_barang[i]}, Jumlah: {self.jumlah_barang[i]}")

    def checkout(self):
        total_harga = sum([harga * jumlah for harga, jumlah in zip(self.harga_barang, self.jumlah_barang)])
        print("Total Harga: Rp", total_harga)
        return total_harga

    def check_order(self):
        if len(self.nama_barang) == len(self.harga_barang) == len(self.jumlah_barang):
            print("Pemesanan sudah benar.")
            print("Transaksi:")
            self.check_keranjang()
        else:
            print("Terdapat kesalahan input data.")

    def total_price(self, total_harga):
        if total_harga > 500000:
            diskon = 0.10
        elif total_harga > 300000:
            diskon = 0.08
        elif total_harga > 200000:
            diskon = 0.05
        else:
            diskon = 0.00

        total_diskon = total_harga * diskon
        total_bayar = total_harga - total_diskon

        print("Total Harga: Rp", total_harga)
        print("Diskon: ", int(diskon * 100), "%")
        print("Total Bayar: Rp", total_bayar)

# Contoh penggunaan
supermarket = Supermarket()

while True:
    print("=== MENU SUPERMARKET ===")
    print("1. Tambah Barang ke Keranjang")
    print("2. Hapus Barang dari Keranjang")
    print("3. Reset Transaksi")
    print("4. Periksa Isi Keranjang")
    print("5. Checkout")
    print("6. Cek Pesanan")
    print("7. Hitung Total Belanja")
    print("0. Keluar")
    choice = input("Pilih menu: ")

    if choice == '1':
        nama = input("Masukkan nama barang: ")
        harga = float(input("Masukkan harga barang: "))
        jumlah = int(input("Masukkan jumlah barang: "))
        supermarket.tambah_barang(nama, harga, jumlah)
    elif choice == '2':
        nama = input("Masukkan nama barang yang ingin dihapus: ")
        supermarket.delete_item(nama)
    elif choice == '3':
        supermarket.reset_transaction()
    elif choice == '4':
        supermarket.check_keranjang()
    elif choice == '5':
        total_harga = supermarket.checkout()
        supermarket.total_price(total_harga)
    elif choice == '6':
        supermarket.check_order()
    elif choice == '7':
        total_harga = supermarket.checkout()
        supermarket.total_price(total_harga)
    elif choice == '0':
        print("Terima kasih telah berbelanja di supermarket!")
        break
    else:
        print("Pilihan tidak valid. Silakan pilih kembali.")
```
